```python
import pandas as pd
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        data = []
        data.append(1)
        data.append(1)
        for i in range(2,n+1):
            data.append(data[i-1] * i)
    return data[n]        
```

```python
def combinations(n, k):
    return factorial(n)/(factorial(k) * factorial(n-k))
```

```python
def one_ticket_probability(numbers):
    possible_outcomes = combinations(49,6)
    number_of_successfull_outcomes = 1 # only one tuple of 6 numbers can win the lottery
    probab = number_of_successfull_outcomes/possible_outcomes
    percentg = probab * 100
    print("Your chances to win the big prize with {} are {:.7f}%".format(numbers, percentg))
    print("In other words, you have 1 in a {:,} chances to win".format(int(possible_outcomes)))
```

```python
one_ticket_probability([1, 1, 1, 1, 1, 1])
```

```python
def check_historical_occurence(user_input, winning_numbers):
    user_input = set(user_input)
    return (winning_numbers.values == user_input).sum()
```

```python
check_historical_occurence([33, 5, 14, 47, 21, 31], winning_numbers)
```

```python
def multi_ticket_probability(n_tickets):
    possible_outcomes = combinations(49, 6)
    successful_outcomes = n_tickets
    probab = successful_outcomes/possible_outcomes
    percentg = probab * 100
    print("Your chances to win the big prize  are {:.7f}%".format(percentg))
    print("In other words, you have {} in a {:,} chances to win".format(n_tickets, int(possible_outcomes)))
```

```python
def probability_less_6(winning_numbers_expected):
    n_combinations_ticket = combinations(6, winning_numbers_expected)
    
    n_combinations_remaining = combinations(43, 6 - winning_numbers_expected)
    successful_outcomes = n_combinations_remaining * n_combinations_ticket
    
    total_possible_outcomes = combinations(49, 6)
    probab = successful_outcomes/total_possible_outcomes
    percentg = probab * 100
    print("Your chances to win a prize are {:.7f}%".format(percentg))
    print("In other words, you have {} in a {:,} chances to win".format(winning_numbers_expected, int(total_possible_outcomes)))
```

```python
probability_less_6(2)
```
